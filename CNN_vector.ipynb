```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
train_datgen = ImageDataGenerator(rescale=1./512)
```

```python
!unzip image.zip
```

```python
#image for folder of input
train_data = train_datgen.flow_from_directory("image",target_size=(512,512),color_mode="rgb",batch_size=1,classes={'USA':'0','UK':'1',\
'ussr':'2','Netherland':'3','India':'4','Poland':'5','egypt':'6','brazil':'7','china':'8','israel':'9','cuba':'10','indonesia':'11','jordan':'12','burma':'13',
},
shuffle=False)
label = train_data.num_classes 
name_of_class = train_data.class_indices
train_set_sample = train_data.samples
print("There is" ,train_set_sample , "images in ",label," classes", name_of_class)
```

```python
import tensorflow as tf 
from keras.applications.resnet50 import ResNet50, preprocess_input,decode_predictions
from keras.layers import *
from keras.models import Model, Sequential
Rsnet50 = ResNet50(weights="imagenet", input_shape=(512,512,3), include_top=False)
trainpool = GlobalAveragePooling2D()(Rsnet50.output)
model = Model(input=Rsnet50.input, output=trainpool)
```

```python
from tqdm import tqdm

def get_features(data,model):
    feats=[]
    label=[]
    count=0
    for i in tqdm(data):
        count+=1
        if (count<= (train_set_sample)): 
            for j in tqdm(i[0]):
              x_in = np.expand_dims(j, axis=0)
              x_in = preprocess_input(x_in)
              features = model.predict(x_in)
              features = features.flatten()
              feats.append(features)
        else: 
           
            break      

    return feats
```

```python
vector = get_features(train_data,model)
```

```python
image_vector = pd.DataFrame(vector)
final_vector = pd.DataFrame(image_vector.values.reshape([14, -1]))
final_vector
```
